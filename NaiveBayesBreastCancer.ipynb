```python
#importação das bibliotecas
import pandas as pd
import numpy as np
```

```python
#Preparando os dados
data=pd.read_csv('data.csv')
data.head()
```

```python
#Definindo os principais prováveis fatores que podem gerar o câncer de mnama
fatores=['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean']
x=data[fatores]
y=data['diagnosis']
```

```python
#Treinando e Testando o modelo
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y = train_test_split(x,y)
```

```python
#colocando o modelo em prática
from sklearn.naive_bayes import GaussianNB

GNB = GaussianNB()
GNB.fit(train_x,train_y)

prediction = GNB.predict(test_x)

```

```python
#avaliação do modelo
from sklearn.metrics import accuracy_score

print(accuracy_score(test_y, prediction))
```
